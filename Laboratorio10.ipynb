```python
import scipy.stats
# Pruebas estadisticas
# La pruebe de Levene (de varianzas)
# HO: las varianzas son iguales
# Ha: las viarianes no son iguales
scipy.stats.levene(df1[df1['Registro de cierre']=='SI']['Avanc financ %'],
                  df1[df1['Registro de cierre']=='NO']['Avanc financ %'],center='mean')

#el p value es menor a 0.05, por lo que la
#Ho se rechaza. Esto significa que las varianzas no son iguales'''
```

```python
# Prueba t
scipy.stats.ttest_ind(df1[df1['Registro de cierre']=='NO']['Avanc financ %'],
                     df1[df1['Registro de cierre']=='SI']['Avanc financ %'],equal_var=False)

#La hipotesis nula es que no hay diferencias significativas en el promedio
#de ambos grupos y la hipotesis alternativa es que si hay diferencias significativas
#en ambos grupos


# el p value es menor a 0.05 por lo que la Ho se rechaza. Hay diferencias en el promedio de
# ambos grupos.
```

```python
# Prueba ANOVA
df1.loc[(df1['Avanc financ %']<=40),'rango%']='40 o menos'
df1.loc[(df1['Avanc financ %']>40) & (df1['Avanc financ %']<=50),'rango%']='Entre 40 y 50'
df1.loc[(df1['Avanc financ %']>50),'rango%']='Mayor a 50'

df1.head(10)
```

```python
# Agrupar la columna
a=df1[df1['rango%']=='40 o menos']['Costo actualizado']
b=df1[df1['rango%']=='Entre 40 y 50']['Costo actualizado']
c=df1[df1['rango%']=='Mayor a 50']['Costo actualizado']

# Ho: los promedios de los grupos son iguales
# Ha: al menos uno de los grupos no tiene un promedio igual
```

```python
estadistico_f,p_value=scipy.stats.f_oneway(a,b,c)
print(estadistico_f,p_value)
# al ser menor a 0.05 (es 0.005) rechazamos la Ho y afirmamos que 
# existen diferencias en el promedio de al menos un grupo
```

```python
# La prueba chi-cuadrado
# Ho: los grupos son independientes
# Ha: los grupos no son independientes

df1.loc[df1['Avanc financ %']<=50,'rango1%']='50 o menos'
df1.loc[df1['Avanc financ %']>50,'rango1%']='Mayor a 50'
df1
```

```python
tab_freq=pd.crosstab(df1['Registro de cierre'],df1['rango1%'])
tab_freq
```

```python
scipy.stats.chi2_contingency(tab_freq,correction=True)
# el p_value es menor a 0.05 por tanto rechazamos la Ho y afirmamos
# que ambos grupos no son independientes.
```
