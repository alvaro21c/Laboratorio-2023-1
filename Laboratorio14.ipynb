```python
df1=df.iloc[0:80,:]
df1.shape
```

```python
# Latitud y longitud de San Francisco
x=37.77
y=-122.42

sanfran=folium.Map(location=[x,y],zoom_start=12)
sanfran
```

```python
df.head()
```

```python
arrestos=folium.map.FeatureGroup()

for lat,lon in zip(df1.Y,df1.X):
  arrestos.add_child(
      folium.features.CircleMarker(
          [lat,lon],
          radius=5,
          color='yellow',
          fill=True,
          fill_color='green',
          fill_opacity=0.5
      )
  )

sanfran.add_child(arrestos)

sanfran
```

```python
from folium import plugins    
df1.head()
```

```python
sanfran=folium.Map(location=[37.77,-122.42],zoom_start=12)
arrestos=plugins.MarkerCluster().add_to(sanfran)

for lat, lon, cat in zip(df1.Y,df1.X,df1.Category):
  folium.Marker(
      location=[lat,lon],
      popup=cat).add_to(arrestos)


sanfran
    
```
