```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('DF.xlsx')
type(df)
```

```python
df.head()
```

```python
df.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace=True)
df
```

```python
df.rename(columns={'OdName':'Pais','AreaName':'Continente','RegName':'Region'},inplace=True)
df
```

```python
df.columns=list(map(str,df.columns))
```

```python
df.columns
```

```python
df.set_index('Pais',inplace=True)
```

```python
df
```

```python
df['Total']=df.sum(axis=1)
df
```

```python
%matplotlib inline
import matplotlib as mpl

mpl.style.use('ggplot')
```

```python
continentes=df.groupby('Continente',axis=0).sum()
continentes
```

```python
continentes['Total'].plot(kind='pie',figsize=(5,6),
                         autopct='%1.1f%%',
                         startangle=90,shadow=True)
plt.title('Inmigracion a Canada por continente,1980-2013')
plt.axis('equal')
plt.show()
```

```python
colores=['yellow','lightskyblue','pink','blue','gold','green']
explode=[0.1,0,0,0,0.1,0.1,0]

continentes['Total'].plot(kind='pie',
                         figsize=(10,6),
                         startangle=90,
                         shadow=True,
                         labels=None,
                         colors=colores,
                         explode=explode,
                         )

plt.title('Grafico 2')
plt.legend(labels=continentes.index,loc='upper left')
plt.show()

```
