```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
type(df)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="sex")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="day")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="time")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="sex",col="day")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="day",col="sex")
plt.show()
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="day",col="sex",size="size")
plt.show()
```

```python
# Grafico distribucion
sns.displot(x="tip",data=df)
plt.show()
```

```python
sns.displot(x="tip",data=df,kind="kde")
plt.show()
```

```python
sns.displot(x="tip",data=df,hue="sex",col="day")
plt.show()
```

```python
sns.catplot(x="tip",y="sex",kind="box",data=df).set_title("Grafico 1")
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",kind="bar",data=df).set_title("Grafico 1")
plt.show()
```
