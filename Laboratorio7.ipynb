```python
# Operaciones dentro de los graficos
import numpy as np
a=np.arange(0,5,0.2)

'''primer grafico : y=x
segundo grafico: y=x^2
tercero grafico: y=x^3'''

plt.plot(a,a,'r-.')
plt.plot(a,a**2,'bo')
plt.plot(a,a**3,'y.')

plt.show()
```
